```python
def random_split_dataset(num_items, train_ratio=0.5, val_ratio=0.25, test_ratio=0.25, seed=42):
    """
    Randomly splits the dataset indices into train/val/test sets with given ratios.

    :param num_items:  Integer, total number of items (images).
    :param train_ratio: Proportion of the dataset to be used for training.
    :param val_ratio:   Proportion of the dataset to be used for validation.
    :param test_ratio:  Proportion of the dataset to be used for testing.
    :param seed:        Random seed for reproducibility.
    :return:            Tuple of three lists: (train_indices, val_indices, test_indices)
    """
    random.seed(seed)
    indices = list(range(num_items))
    random.shuffle(indices)

    total_len = len(indices)
    train_size = int(train_ratio * total_len)
    val_size = int(val_ratio * total_len)
    # The remainder goes to test, but let's calculate it explicitly:
    # test_size = int(test_ratio * total_len)  # If needed

    train_indices_ = indices[:train_size]
    val_indices_ = indices[train_size:train_size + val_size]
    test_indices_ = indices[train_size + val_size:]

    return train_indices_, val_indices_, test_indices_
```

```python
# Function to split dataset indices into train, validation, and test sets
def split_dataset_indices(total_samples, train_frac, val_frac, test_frac, random_seed):
    import random
    random.seed(random_seed)

    indices = list(range(total_samples))
    random.shuffle(indices)

    train_cutoff = int(train_frac * total_samples)
    val_cutoff = train_cutoff + int(val_frac * total_samples)

    train_set = indices[:train_cutoff]
    val_set = indices[train_cutoff:val_cutoff]
    test_set = indices[val_cutoff:]

    return train_set, val_set, test_set

# Generate dataset splits
train_set, val_set, test_set = split_dataset_indices(
    total_samples=len(image_paths),
    train_frac=0.5,
    val_frac=0.25,
    test_frac=0.25,
    random_seed=42
)

# Display dataset distribution
print("\n--- Dataset Split Summary (Train: 50%, Val: 25%, Test: 25%) ---")
print(f"Training samples: {len(train_set)}")
print(f"Validation samples: {len(val_set)}")
print(f"Testing samples: {len(test_set)}")

# Calculate and display the proportion of each set
total_count = len(train_set) + len(val_set) + len(test_set)
print("\nProportion of each dataset split:")
print(f"Training set: {len(train_set) / total_count:.2%}")
print(f"Validation set: {len(val_set) / total_count:.2%}")
print(f"Testing set: {len(test_set) / total_count:.2%}")
```

```python

class FlowerImageDataset(Dataset):
    """
    Custom PyTorch Dataset for handling the Flowers102 dataset.

    Attributes:
        sample_indices (list): A list of indices specifying dataset samples.
        category_labels (array): A numpy array of flower category labels (1-based indexing).
        file_paths (list): A sorted list of image file paths.
        augmentation (callable, optional): Transformations to be applied to each image.
    """

    def __init__(self, sample_indices, category_labels, file_paths, augmentation=None):
        """
        Args:
            sample_indices (list): Indices of selected images for this dataset partition.
            category_labels (array): Flower class labels (1-based indexing) for all images.
            file_paths (list): Complete paths to all images.
            augmentation (callable, optional): Function or transform applied to input images.
        """
        self.sample_indices = sample_indices
        self.category_labels = category_labels
        self.file_paths = file_paths
        self.augmentation = augmentation

    def __len__(self):
        return len(self.sample_indices)

    def __getitem__(self, index):
        # Retrieve actual image index from the dataset
        actual_index = self.sample_indices[index]
        img_path = self.file_paths[actual_index]
        class_label = int(self.category_labels[actual_index]) - 1  # Convert to 0-based indexing

        if os.path.exists(img_path):
            # Open image and convert it to RGB
            img = Image.open(img_path).convert("RGB")
        else:
            # Handle missing image files by replacing them with a blank black image
            print(f"Warning: Missing file at {img_path}. Replacing with a black placeholder image.")
            img = Image.new("RGB", (224, 224), color="black")
            class_label = 0  # Assign a dummy label

        # Apply transformation if provided
        if self.augmentation:
            img = self.augmentation(img)

        return img, class_label
```

```python
import torch

def train_neural_net(model, loss_function, model_optimizer, training_loader, validation_loader, num_epochs):
    """
    Trains and evaluates a neural network model.

    Args:
        model (torch.nn.Module): The neural network to be trained.
        loss_function (torch.nn.Module): The loss function used for optimization.
        model_optimizer (torch.optim.Optimizer): The optimizer for updating model parameters.
        training_loader (torch.utils.data.DataLoader): DataLoader for the training set.
        validation_loader (torch.utils.data.DataLoader): DataLoader for the validation set.
        num_epochs (int): Number of training iterations.

    Returns:
        tuple: Training losses, validation losses, training accuracies, and validation accuracies.
    """

    training_losses, validation_losses = [], []
    training_accuracies, validation_accuracies = [], []

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        epoch_train_loss, correct_train, total_train = 0.0, 0, 0

        for batch_images, batch_labels in training_loader:
            batch_images, batch_labels = batch_images.to(compute_device), batch_labels.to(compute_device)

            # Forward and backward pass
            model_optimizer.zero_grad()
            output_predictions = model(batch_images)
            batch_loss = loss_function(output_predictions, batch_labels)
            batch_loss.backward()
            model_optimizer.step()

            epoch_train_loss += batch_loss.item() * batch_images.size(0)
            _, predicted_labels = torch.max(output_predictions, 1)
            correct_train += (predicted_labels == batch_labels).sum().item()
            total_train += batch_labels.size(0)

        training_losses.append(epoch_train_loss / len(training_loader.dataset))
        training_accuracies.append(correct_train / total_train * 100)

        # Validation phase
        model.eval()
        epoch_val_loss, correct_val, total_val = 0.0, 0, 0
        with torch.no_grad():
            for batch_images, batch_labels in validation_loader:
                batch_images, batch_labels = batch_images.to(compute_device), batch_labels.to(compute_device)
                output_predictions = model(batch_images)
                batch_loss = loss_function(output_predictions, batch_labels)

                epoch_val_loss += batch_loss.item() * batch_images.size(0)
                _, predicted_labels = torch.max(output_predictions, 1)
                correct_val += (predicted_labels == batch_labels).sum().item()
                total_val += batch_labels.size(0)

        validation_losses.append(epoch_val_loss / len(validation_loader.dataset))
        validation_accuracies.append(correct_val / total_val * 100)

        print(f"Epoch [{epoch+1}/{num_epochs}], "
              f"Train Loss: {training_losses[-1]:.4f}, Train Acc: {training_accuracies[-1]:.2f}%, "
              f"Val Loss: {validation_losses[-1]:.4f}, Val Acc: {validation_accuracies[-1]:.2f}%")

    return training_losses, validation_losses, training_accuracies, validation_accuracies
```

```python
import torch

def assess_model(model_instance, data_loader):
    """
    Evaluates the model's accuracy on a given dataset.

    Args:
        model_instance (torch.nn.Module): The trained model.
        data_loader (DataLoader): DataLoader containing test/validation samples.

    Returns:
        float: Model accuracy as a percentage.
    """
    model_instance.eval()
    accurate_predictions, sample_count = 0, 0

    with torch.no_grad():
        for batch_inputs, batch_targets in data_loader:
            batch_inputs, batch_targets = batch_inputs.to(compute_device), batch_targets.to(compute_device)
            model_outputs = model_instance(batch_inputs)
            _, predicted_classes = torch.max(model_outputs, 1)

            accurate_predictions += (predicted_classes == batch_targets).sum().item()
            sample_count += batch_targets.size(0)

    return accurate_predictions / sample_count * 100
```

```python
def visualize_predictions(model_instance, data_loader, class_names, num_examples=8):
    """
    Visualizes predictions made by the model on a grid of images.

    Args:
        model_instance (torch.nn.Module): Trained model for prediction.
        data_loader (DataLoader): DataLoader containing the dataset.
        class_names (list): List of class names corresponding to labels.
        num_examples (int): Number of example images to display in the grid.
    """
    model_instance.eval()  # Set the model to evaluation mode
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Get a batch of images and labels
    images, labels = next(iter(data_loader))
    images, labels = images[:num_examples], labels[:num_examples]
    images = images.to(device)

    # Get predictions
    with torch.no_grad():
        outputs = model_instance(images)
        _, preds = torch.max(outputs, 1)

    images = images.cpu()  # Move images back to CPU for visualization
    labels = labels.cpu()
    preds = preds.cpu()

    # Plot the images with predictions
    fig, axs = plt.subplots(2, 4, figsize=(12, 6))  # 2 rows, 4 columns grid
    axs = axs.flatten()  # Flatten to iterate over all subplots

    for i in range(num_examples):
        image = images[i].permute(1, 2, 0).numpy()  # Convert tensor to NumPy format
        image = (image - image.min()) / (image.max() - image.min())  # Normalize for display
        axs[i].imshow(image)
        axs[i].set_title(f"Pred: {class_names[preds[i]]}\nTrue: {class_names[labels[i]]}")
        axs[i].axis("off")

    # Hide unused subplots if any
    for j in range(num_examples, len(axs)):
        axs[j].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
import torch
import torch.nn as nn

class CustomYOLOv5Classifier(nn.Module):
    def __init__(self, pretrained_yolo_model, output_classes=102):
        super().__init__()
        # Extract layers up to index 9 (excluding final layers) from the pretrained YOLO model
        self.feature_extractor = pretrained_yolo_model.model.model.model[:10]  # Up to SPPF
        self.pooling_layer = nn.AdaptiveAvgPool2d((1, 1))  # Global average pooling layer
        self.dense1 = nn.Linear(204800, 1024)  # Fully connected layer 1
        self.dense2 = nn.Linear(1024, 512)  # Added additional dense layer
        self.dense3 = nn.Linear(512, output_classes)  # Final output layer

    def forward(self, input_tensor):
        feature_map = self.feature_extractor(input_tensor)  # Get features from YOLOv5
        # feature_map = self.pooling_layer(feature_map)  # Optional: Pooling to reduce spatial dimensions
        flattened = torch.flatten(feature_map, 1)  # Flatten feature map to (batch_size, 204800)
        dense_output1 = self.dense1(flattened)  # Pass through first dense layer
        activated_output1 = torch.relu(dense_output1)  # Apply ReLU activation
        dense_output2 = self.dense2(activated_output1)  # Pass through second dense layer
        activated_output2 = torch.relu(dense_output2)  # Apply ReLU activation
        final_output = self.dense3(activated_output2)  # Pass through final output layer
        return final_output
```

## Training

### Train

```python
import torch
import torch.nn as nn

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(classifier_model.parameters(), lr=0.001)

# Modified training function
def train_classifier(model, train_data_loader, validation_data_loader, num_epochs):
    # Lists to store the metrics
    training_losses = []
    validation_losses = []
    validation_accuracies = []
    training_accuracies = []

    for epoch in range(num_epochs):
        # Model training phase
        model.train()
        epoch_train_loss = 0.0
        correct_train_predictions, total_train_samples = 0, 0

        for input_images, target_labels in train_data_loader:
            input_images, target_labels = input_images.to(device), target_labels.to(device)

            optimizer.zero_grad()
            model_outputs = model(input_images)
            loss_value = criterion(model_outputs, target_labels)
            loss_value.backward()
            optimizer.step()
            epoch_train_loss += loss_value.item()

            _, predicted_train_labels = torch.max(model_outputs, 1)
            total_train_samples += target_labels.size(0)
            correct_train_predictions += (predicted_train_labels == target_labels).sum().item()

        avg_training_loss = epoch_train_loss / len(train_data_loader)
        training_accuracy = 100 * correct_train_predictions / total_train_samples
        training_losses.append(avg_training_loss)  # Track the training loss
        training_accuracies.append(training_accuracy)  # Track the training accuracy
        print(f"Epoch [{epoch+1}/{num_epochs}], Training Loss: {avg_training_loss:.4f}, Training Accuracy: {training_accuracy:.2f}%")

        # Model validation phase
        model.eval()
        epoch_val_loss = 0.0
        correct_val_predictions, total_val_samples = 0, 0

        with torch.no_grad():
            for input_images, target_labels in validation_data_loader:
                input_images, target_labels = input_images.to(device), target_labels.to(device)
                model_outputs = model(input_images)
                loss_value = criterion(model_outputs, target_labels)
                epoch_val_loss += loss_value.item()

                _, predicted_val_labels = torch.max(model_outputs, 1)
                total_val_samples += target_labels.size(0)
                correct_val_predictions += (predicted_val_labels == target_labels).sum().item()

        avg_validation_loss = epoch_val_loss / len(validation_data_loader)
        validation_accuracy = 100 * correct_val_predictions / total_val_samples
        validation_losses.append(avg_validation_loss)  # Track the validation loss
        validation_accuracies.append(validation_accuracy)  # Track the validation accuracy
        print(f"Validation Loss: {avg_validation_loss:.4f}, Validation Accuracy: {validation_accuracy:.2f}%")

    return training_losses, validation_losses, training_accuracies, validation_accuracies
```

```python
# Train the classifier
train_losses, val_losses,train_accuracies, val_accuracies = train_classifier(classifier_model, train_loader, validation_loader, num_epochs=5)
```

### Eval

```python
# Modified testing function
def evaluate_classifier(model, test_data_loader):
    model.eval()
    correct_predictions, total_samples = 0, 0
    with torch.no_grad():
        for input_images, target_labels in test_data_loader:
            input_images, target_labels = input_images.to(device), target_labels.to(device)
            model_outputs = model(input_images)
            _, predicted_labels = torch.max(model_outputs, 1)
            total_samples += target_labels.size(0)
            correct_predictions += (predicted_labels == target_labels).sum().item()

    accuracy = 100 * correct_predictions / total_samples
    print(f"Test Accuracy: {accuracy:.2f}%")

# Evaluate the classifier
evaluate_classifier(classifier_model, test_loader)
```
